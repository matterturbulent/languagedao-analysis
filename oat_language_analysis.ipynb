```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv("oat_cleaned_data.csv")
print(df.head(10))
print("\n")

# Create a new series containing each instance of someone speaking a particular language
language_list = []
for row in df["language"]:
    if "," in row:
        split_row = row.split(",")
        for item in split_row:
            language_list.append(item)
    else:
        language_list.append(row)     
language_list_corrected = pd.DataFrame({"language":language_list})

# Apply title capitalization to all the language names 
final_language_list = language_list_corrected["language"].str.strip()
final_language_list = final_language_list.str.title()
 
# Write the dataframe as a new CSV file      
final_language_list.to_csv("final_language_list.csv")

# Calculate proportions of languages spoken, write to a new CSV file
counts = pd.DataFrame()
counts["language_count"] = final_language_list.value_counts()
counts["percent"] = counts["language_count"] / counts["language_count"].sum()
counts["percent"] = counts["percent"].apply(lambda x: "{:.2%}".format(x))
counts.to_csv("oat_language_counts.csv")
print(counts)

```
